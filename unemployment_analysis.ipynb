```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x='Region',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
top10=df.sort_values(by='2021',ascending=False).head(10)
sns.barplot(x='Country Name',y='2021',data=top10)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
top10=df.sort_values(by='2021',ascending=False).head(10)
sns.barplot(x='Country Name',y='2021',data=top10)
plt.xticks(rotation=90)
plt.title("Top 10 Countries With Highest Unemployment Rate In 2021")
plt.show()
```

```python
top10lowest=df.sort_values(by='2021',ascending=True).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='Country Name',y='2021',data=top10lowest)
plt.xticks(rotation=90)
plt.title("Top 10 Countries With Lowest Unemployment Rate In 2021")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.select_dtypes(include='number').corr())
plt.title("Correlation Heatmap")
plt.show()
```

```python
highest=df.loc[df['2021'].idxmax()]
print(highest)
```

```python
lowest=df.loc[df['2021'].idxmin()]
print(lowest)
```

```python
plt.figure(figsize=(12,6))
top10=df.sort_values(by='2021',ascending=False).head(10)
sns.barplot(x='Country Name',y='2021',data=top10)
plt.xticks(rotation=90)
plt.title("Top 10 Countries With Highest Unemployment Rate In 2021")
plt.savefig("highest_unemployment.png")
plt.show()
```

```python
plt.figure(figsize=(12,6))
top10lowest=df.sort_values(by='2021',ascending=True).head(10)
sns.barplot(x='Country Name',y='2021',data=top10lowest)
plt.xticks(rotation=90)
plt.title("Top 10 Countries With Lowest Unemployment Rate In 2021")
plt.savefig("lowest_unemployment.png")
plt.show()
```
