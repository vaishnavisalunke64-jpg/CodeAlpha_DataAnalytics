```python
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
reviews=["This product is amazing and very useful","I really liked the quality of this item","Worst purchase I have ever made","This is not good and I am disappointed","Excellent product and fast delivery","Very bad experience with this product","I am extremely happy with this purchase","The quality is poor and terrible","Nice design and good performance","I will never buy this again","This product works perfectly and I am satisfied","Very disappointing quality and waste of money","I love the features and performance","Bad packaging and late delivery","Amazing experience and excellent support","The product stopped working after one day","Very comfortable and easy to use","Terrible customer service experience","I am happy with the product quality","Not satisfied with the overall performance"]
df=pd.DataFrame(reviews,columns=["Review"])
sentiment=[]
polarity_score=[]
for i in df["Review"]:
    analysis=TextBlob(i)
    score=analysis.sentiment.polarity
    polarity_score.append(score)
    if score>0:
        sentiment.append("Positive")
    elif score<0:
        sentiment.append("Negative")
    else:
        sentiment.append("Neutral")
df["Polarity Score"]=polarity_score
df["Sentiment"]=sentiment
print(df)
positive_count=df[df["Sentiment"]=="Positive"].shape[0]
negative_count=df[df["Sentiment"]=="Negative"].shape[0]
neutral_count=df[df["Sentiment"]=="Neutral"].shape[0]
print("Positive Reviews:",positive_count)
print("Negative Reviews:",negative_count)
print("Neutral Reviews:",neutral_count)
plt.figure(figsize=(8,5))
sns.countplot(x="Sentiment",data=df)
plt.title("Sentiment Analysis Of Product Reviews")
plt.xlabel("Sentiment Type")
plt.ylabel("Number Of Reviews")
plt.show()
```
