```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://quotes.toscrape.com"
```

```python
response=requests.get(url)
```

```python
soup=BeautifulSoup(response.text,'html.parser')
```

```python
quotes=soup.find_all('span',class_='text')
```

```python
for quote in quotes:
    print(quote.text)
```
